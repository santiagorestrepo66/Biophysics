This code allows retrieving the experimental B-factors of the alpha carbon backbone of a biomolecule from its corresponding PDB file.

DESCRIPTIONS
-------------

**File type (PDB or CIF):** The crystal structure file extension.

**File name:** The name of the file containing the crystal structure (do not include the extension).

**Number of residues:** This output tells you the number of residues of the biomolecule which experimental B-factors you are trying to retrieve.



```python
# LIBRARIES AND FUNCTIONS

# Libraries
print('\033[37m' + 'Importing libraries...'  + '\033[0m')
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
import numpy as np
import os

# String substraction
print('\033[37m' + 'Defining string substraction function...'  + '\033[0m')
def substract(a, b):                              
    return "".join(a.rsplit(b))

print('\033[1m' + 'Done' + '\033[0m')
```

```python
# EXPERIMENTAL B-FACTORS

# Enable dynamic graphs mode
%matplotlib nbagg

# File type
FileType = 0
while FileType == 0:
    FileType = input('\033[1m' + 'File type (PDB or CIF): ' + '\033[0')
    if FileType not in {'PDB', 'CIF'}:
        FileType = 0
        print('Invalid file type')
    elif FileType in {'PDB', 'CIF'}:
        FileType = str(FileType)
        
# File directory
FileName = 0
while FileName == 0:
    path = 0
    FileName = input('\033[1m' + 'File name: ' + '\033[0')
    if FileName.startswith('C') == True:
        FileName = 0
        print('This file does not contain accurate B-factors')
    else:
        if FileType == 'PDB':
            FileName = FileName + '.pdb'
        elif FileType == 'CIF':
            FileName = FileName + '.cif'
        path  = os.path.join(r'C:\Users\santi\Desktop\2019\Biophysics\PDBs', FileName)
        if Path(path).is_file():
            if FileType == 'PDB':
                MoleculeName = substract(FileName,'.pdb')
            elif FileType == 'CIF':
                MoleculeName = substract(FileName,'.cif')
        else:
            FileName = 0
            print('File not found')
        
# Experimental B-factors retrieval
print('\033[37m' + 'Retrieving alpha carbon backbone experimental B-factors...'  + '\033[0m')
N = 0
BFactorsList = []
with open(path) as file:
    for line in file:
        if line[:4] == 'ATOM':
            if FileType == 'PDB':
                if line[13:15] == 'CA':
                    N += 1
                    BFactorsList.append(line[60:66])
            elif FileType == 'CIF':
                if line[15:17] == 'CA':
                    N += 1
                    BFactorsList.append(line[60:66])
file.closed
print('Number of residues: ' + str(N))
BFactors = np.zeros((N))
for i in range(0,N):
    BFactors[i] = float(BFactorsList[i])
    
# Experimental B-factors graph
print('\033[37m' + 'Graphing experimental B-factors vs. residue index...'  + '\033[0m')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 6
fig, ax = plt.subplots()
ax.set(xlabel = 'Residue index', ylabel = 'Experimental B-factors (angstroms^2)')
i = np.arange(1, N + 1, 1)
ax.scatter(i, BFactors, c = 'black', s = 2)
ax.plot(i, BFactors, c = 'gray', linewidth = 0.5, linestyle = '-')
Min = np.amin(BFactors)
Max = np.amax(BFactors)
Range = Max - Min
ax.set_ylim(Min - 0.05*Range, Max + 0.05*Range)
plt.show()

print('\033[1m' + 'Done' + '\033[0m')
```

```python
# SAVE FILES

# Experimental B-factors dataframe
print('\033[37m' + 'Creating experimental B-factors dataframe...'  + '\033[0m')
df = pd.DataFrame({'B-factors (angstroms^2)': BFactors})
df.index = np.arange(1, len(df) + 1)
df.index.name = 'Node'

# File saving
print('\033[37m' + 'Saving experimental B-factors...'  + '\033[0m')
df.to_csv(MoleculeName + '_CG_experimentalB-factors.csv')

print('\033[1m' + 'Done' + '\033[0m')
```
